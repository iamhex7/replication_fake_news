# Discrepancies Between Pre-analysis Plan and Article

This section compares the pre-analysis plan, the published article, and the replication outputs for **Figure 1, Figure 3, Figure 4, Table 2, and Table 3**. I organize the discussion according to three dimensions: **research design**, **variables and outcomes**, and **empirical strategy**.


##  Research Design

Overall, the replication follows the broad research design of the article. The main research question is whether respondents evaluate the truthfulness of news differently depending on whether the news is aligned with or opposed to their political preferences. Across the figures and tables, the central comparison is between **pro-party**, **anti-party**, **neutral**, **true**, and **fake** news. This is consistent with the article’s general design and its focus on motivated reasoning in political news evaluation.

### Figure 1

**Discrepancy signal:** The original article uses a distributional comparison, while the replication depends on reconstructed treatment-category definitions.

Figure 1 in the article presents cumulative distribution functions of beliefs about $P(\text{True News})$ separately for pro-party and anti-party news. This design emphasizes the full distribution of beliefs rather than only average differences. The replication follows the same broad idea by plotting ECDFs for pro-party and anti-party news, but it relies on reconstructed grouping variables from the cleaned dataset. Therefore, while the visual design is similar, the exact classification of observations depends on how variables such as `good_news`, `bad_news`, `pro_party`, and `anti_party` are operationalized in the replication.

### Figure 3

**Discrepancy signal:** The neutral-news sample sizes differ from the original output, suggesting a mismatch in sample construction.

Figure 3 introduces a more noticeable discrepancy in research design. The article’s original figure compares demeaned beliefs across six groups: anti-party fake news, anti-party true news, neutral fake news, neutral true news, pro-party fake news, and pro-party true news. The replication follows this same six-category structure, but the sample sizes differ for the neutral-news categories. In the original figure, the neutral categories have much larger sample sizes, while the replication produces smaller neutral groups. This suggests that the replication’s sample restriction may not fully match the article’s original inclusion criteria for neutral news.

### Figure 4

**Discrepancy signal:** Some topic-specific effects cannot be estimated after demeaning because variables become collinear or have no remaining variation.

Figure 4 follows the article’s general design by examining heterogeneity across topics. The article presents topic-specific estimates of motivated reasoning, such as pro-party news interacted with issue categories. The replication attempts to reproduce this by using `good_topic` indicators and estimating their effects on $P(\text{True})$. However, some topic variables are dropped in the replication because they become collinear or have zero variance after fixed-effect demeaning. This creates a discrepancy between the article’s intended topic-by-topic design and the estimable design in the replication.

### Table 2

**Discrepancy signal:** The core design is close to the article, but differences may arise from manual construction of variables and fixed effects.

Table 2 is closest to the article’s main research design. Both the article and replication estimate treatment effects of pro-party news, anti-party news, true news, and related interactions under different fixed-effect and control specifications. The replicated point estimates are mostly close to the original table, especially for the main pro-party news effect. This suggests that the core research design is broadly reproduced.

### Table 3

**Discrepancy signal:** The replication sample size differs substantially from the article, which affects comparability.

For Table 3, the discrepancy is larger. The original table uses 4,085 observations, while the replication output uses 7,902 observations. This indicates that the replication sample is not the same as the article’s sample for the belief-updating regressions. Since Table 3 concerns changes in beliefs, the article likely applies a more restrictive sample definition, while the replication appears to use the broader political-news sample. This is a major research-design discrepancy because changing the sample can directly affect estimated coefficients, standard errors, and $R^2$.

### Summary of Research Design Discrepancies

| Output | Main research-design discrepancy | Why it matters |
|---|---|---|
| Figure 1 | Similar CDF design, but category construction may differ | Grouping definitions affect the plotted distributions |
| Figure 3 | Neutral-news sample sizes differ | Suggests sample restriction mismatch |
| Figure 4 | Some topic effects cannot be estimated after demeaning | Limits direct comparison of topic heterogeneity |
| Table 2 | Broad design is similar | Main treatment-effect logic is preserved |
| Table 3 | Observation count differs sharply | Major limitation for exact replication |


## 3.2 Variables and Outcomes

The main outcome variable throughout the article is respondents’ belief about the probability that a news item is true, usually measured as $P(\text{True})$ or $P(\text{True News})$. The replication uses the same conceptual outcome, `prob_true`, for Figures 1, 3, 4, and Table 2. This is broadly consistent with the article.

### Figure 1

**Discrepancy signal:** The outcome is conceptually the same, but the pro-party and anti-party categories may not be constructed exactly the same way.

For Figure 1, both the article and replication use beliefs about $P(\text{True News})$ as the outcome. The article displays the entire distribution of this outcome using CDFs for pro-party and anti-party news. The replication follows the same approach by computing ECDFs from `prob_true`. The main discrepancy is not the outcome itself, but the construction of the pro-party and anti-party categories. The replication creates the news grouping from variables such as `good_news` and `bad_news`, while the article’s labels are framed directly as “Pro-Party News” and “Anti-Party News.” If these variables are not perfectly equivalent, the replicated figure may group observations differently from the article.

### Figure 3

**Discrepancy signal:** The replication uses a simplified demeaning procedure, which may differ from the article’s fixed-effect-adjusted demeaning.

For Figure 3, the outcome is again $P(\text{True})$, but the article reports it after demeaning. The replication reproduces this idea by subtracting the overall mean from each group mean. However, this is likely not exactly the same demeaning procedure as in the article. The article’s y-axis is labeled “$P(\text{True})$, Demeaned,” which likely means the outcome was residualized with respect to fixed effects or comparison groups, not simply demeaned by the overall sample mean. The replication’s demeaned mean is calculated as:

$$
\text{group mean} - \text{overall mean}.
$$

This creates a discrepancy because the article’s demeaning may account for experimental structure, question fixed effects, round fixed effects, or subject-level variation, while the replication uses a simpler overall mean adjustment.

There is also a discrepancy in the neutral-news variables in Figure 3. In the original figure, neutral fake and neutral true news have sample sizes of about 1,310 and 1,340 respectively. In the replication, the neutral fake and neutral true news categories are much smaller. This suggests that the replication may be excluding some neutral observations, possibly because it filters first on variables that do not capture all neutral news items in the same way as the article.

### Figure 4

**Discrepancy signal:** The topic variables in the replication are not fully mapped to the article’s topic labels, and some are dropped after demeaning.

For Figure 4, the article uses topic-specific interaction variables, such as pro-party news interacted with climate, race, mobility, refugees, gender, media, and other topics. The replication uses variables named `good_topic1`, `good_topic2`, and so on, and then maps them to topic labels. This creates a variable-labeling discrepancy. Some variables remain labeled generically, such as `good_topic14` and `good_topic15`, rather than being fully mapped to article topic names. In addition, the replication drops some topic variables after demeaning because they have zero variance or are collinear. Therefore, the set of topic regressors shown in the replication is not exactly the same as the article’s full set of topic-specific estimates.

### Table 2

**Discrepancy signal:** The main variables are similar, but constructed interaction variables may not exactly match the article’s definitions.

For Table 2, the variables are largely consistent with the article. The replication includes `pro_party`, `anti_party`, `true_news`, and the interaction between partisanship strength and pro-party news. The signs and magnitudes of the main coefficients are also similar to the original table. However, there are some discrepancies in implementation. For example, the replication constructs `pro_party_str` as `pro_party * abs_net_party`, while the exact article variable may have been created using a more specific measure of partisan strength. This may explain why some standard errors in the replicated table become unstable.

### Table 3

**Discrepancy signal:** The dependent variable and sample for belief updating do not fully match the article.

For Table 3, the main outcome changes from perceived truthfulness to belief updating, measured in the replication as `change_guess_message`. This matches the general idea of Table 3, which studies changes in beliefs. However, the replication’s sample and model structure do not fully match the original. The original Table 3 reports 4,085 observations, while the replication reports 7,902. Also, the original Table 3 has much higher $R^2$ values, around 0.28 and 0.45, while the replication reports much lower $R^2$ values, around 0.01 and 0.22. This suggests that the replication may not be including the same fixed effects or may be using a different dependent-variable sample.

| Figure 1 | Grouping variables may differ from article labels | Affects which observations enter each CDF |
| Figure 3 | Demeaning procedure may be simplified | Affects group means and confidence intervals |
| Figure 4 | Topic labels and estimable topic indicators differ | Limits one-to-one comparison across topics |
| Table 2 | Interaction variable construction may differ | Can affect coefficient magnitudes and SEs |
| Table 3 | Belief-updating outcome sample differs | Affects coefficient estimates and model fit |


## 3.3 Empirical Strategy

The article’s empirical strategy combines descriptive figures, coefficient plots, and regression tables. The replication follows the same broad strategy, but there are several implementation differences across the figures and tables.

### Figure 1

**Discrepancy signal:** The ECDF strategy is appropriate, but the replication depends on manually reconstructed sample restrictions.

For Figure 1, the empirical strategy is descriptive. The article plots CDFs of $P(\text{True News})$ for pro-party and anti-party news. The replication uses an ECDF function to reproduce this structure. This is a reasonable replication choice because CDFs are nonparametric and do not require regression assumptions. The main discrepancy is that the replication depends on manually constructed sample restrictions, such as keeping political-news observations and excluding observations that do not fall into the reconstructed pro-party or anti-party categories. If the original article used a slightly different Stata sample condition, then the replicated CDFs may not match exactly.

### Figure 3

**Discrepancy signal:** The replication uses group means with overall demeaning rather than the article’s likely fixed-effect residualization.

For Figure 3, the article’s strategy appears to be a fixed-effect-adjusted comparison of demeaned beliefs across news categories. The replication instead computes group means and subtracts the overall mean. This is simpler and easier to implement, but it may not reproduce the exact residualization procedure used in the article. The discrepancy matters because fixed-effect demeaning can change group averages if different categories are unevenly distributed across rounds, topics, or subjects. Therefore, the replication captures the qualitative comparison across categories, but may not reproduce the exact article estimates.

### Figure 4

**Discrepancy signal:** Manual fixed-effect demeaning creates collinearity problems that require dropping variables or using a pseudo-inverse.

For Figure 4, the replication uses a manual fixed-effects strategy. It demeans both the dependent variable and the regressors within `round_number` and `topic_id` groups, then estimates OLS without an intercept. This is an attempt to replicate fixed effects without creating many dummy variables. However, this strategy creates collinearity problems. Several `good_topic` variables become zero-variance or perfectly collinear after demeaning, so the replication drops them before estimating cluster-robust standard errors. The article likely used Stata or another econometric package that automatically absorbs fixed effects and drops collinear terms. As a result, the replication’s Figure 4 is not perfectly comparable to the original because some topic-specific coefficients are omitted or renamed.

### Table 2

**Discrepancy signal:** Point estimates are close, but some replicated standard errors are unstable under more saturated specifications.

For Table 2, the empirical strategy is closest to the article. The replication estimates six OLS specifications with different combinations of controls and fixed effects. The point estimates are largely consistent with the original Table 2. For example, the pro-party news coefficient is approximately 0.092 in column (1), close to the article’s reported value, and the signs of anti-party news and true news also match the article. However, the replicated standard errors in some columns are problematic. In particular, the standard errors for columns with subject fixed effects or interacted fixed effects become extremely large in the replication output. This suggests that the regression design matrix may be nearly singular or that the cluster-robust covariance matrix is unstable. The original table reports small and precise standard errors, so this is an important empirical-strategy discrepancy.

### Table 3

**Discrepancy signal:** The replication table labels fixed effects as included, but the implemented formulas appear too simple to reproduce the article’s fixed-effect specification.

For Table 3, the empirical-strategy discrepancy is more substantial. The original Table 3 includes question fixed effects, round fixed effects, and subject fixed effects in all six columns. The replication table labels these fixed effects as “Yes,” but the actual formulas shown in the notebook are simple regressions such as:

$$
\text{change\_guess\_message} \sim \text{pro\_party},
$$

or

$$
\text{change\_guess\_message} \sim \text{pro\_party} + \text{polarizing} + \text{prob\_true}.
$$

These formulas do not appear to include the fixed effects explicitly, such as `C(round_number)`, `C(topic_id)`, or `C(code)`. This likely explains why the replicated $R^2$ values are much lower than those in the original table. The article’s Table 3 reports $R^2$ values around 0.28 and 0.45, while the replication reports values around 0.01 and 0.22. The difference suggests that the replication’s Table 3 does not fully implement the article’s fixed-effect specification.

The sample size discrepancy in Table 3 reinforces this point. The article reports 4,085 observations, while the replication reports 7,902 observations. This suggests that the original Table 3 applies a more restrictive sample, likely based on observations with valid belief-updating measures or message exposure, while the replication uses all political-news observations. Because both the sample and fixed effects differ, the replicated Table 3 should be interpreted as a partial replication rather than an exact reproduction.

### Summary of Empirical Strategy Discrepancies

| Output | Main empirical-strategy discrepancy | Why it matters |
|---|---|---|
| Figure 1 | Manual sample restriction for ECDFs | May shift distributional comparisons |
| Figure 3 | Simplified demeaning | May change adjusted group means |
| Figure 4 | Manual demeaning creates singularity and dropped variables | Reduces comparability of topic coefficients |
| Table 2 | Cluster-robust SEs become unstable in saturated models | Affects statistical inference |
| Table 3 | Fixed effects and sample restrictions not fully implemented | Major source of mismatch in observations and $R^2$ |

Overall, the empirical strategy is directionally similar to the article but differs in important technical details. The replication captures the broad logic of motivated-reasoning regressions, but some results differ because of simplified demeaning, different sample restrictions, manual handling of fixed effects, unstable cluster-robust standard errors, and incomplete implementation of fixed effects in Table 3. These discrepancies are important because they affect direct comparability between the replication outputs and the published article, especially for Figure 3, Figure 4, and Table 3.
